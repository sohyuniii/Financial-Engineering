```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
x
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig=plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)
fig
```

```python
fig=plt.figure()
plt.plot(x,y1,".")
plt.plot(x,y2,"--")
fig
```

### scatter plot 

```python
rng = np.random.RandomState(0)
```

```python
x = rng.randn(100)
y = rng.randn(100)
color = rng.rand(100)
size = 1000*rng.rand(100)

plt.scatter(x,y,c=color,s=size,alpha=0.5,cmap="autumn")
plt.colorbar()
plt.show()
```

### contour plot 

```python
def f(x,y):
    z = x**2 + y**2
    return z
```

```python
x = np.linspace(-2,2,20)
y = np.linspace(-4,4,30)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
plt.contour(X,Y,Z,20,cmap="winter")
plt.show()
```

```python
plt.contour(X,Y,Z,100,cmap="winter")
plt.show()
```

```python
plt.imshow(Z, extent=[-2,2,-4,4], origin = "lower",cmap="RdGy")
plt.colorbar()
plt.show()
```

```python
# two plots at the same time
contours = plt.contour(X, Y, Z, 3)
plt.clabel(contours, inline=True, fontsize=25)
plt.imshow(Z, extent=[-2,2,-4,4])
plt.show()
```
